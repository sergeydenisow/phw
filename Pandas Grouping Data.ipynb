## Задание №1.

```python
import pandas as pd
```

```python
def classify(rating):
    """
    
    Функция классифицирует фильмы по трём категориям в зависимости от величины рейтинга:
    
    2 или меньше - низкий рейтинг
    больше 2, но меньше или равно 4 - средний рейтинг
    выше 4 - высокий рейтинг.
    
    
    """
    score = ["низкий рейтинг", "средний рейтинг", "высокий рейтинг"]
    if rating <= 2:
        return score[0]
    elif rating <= 4:
        return score[1]
    else: return score[2]
```

```python
# Путь к файлам данных.
path_1 = '../ml-latest-small/'

# Загружаем данные по рейтинга.
ratings = pd.read_csv(path_1 + 'ratings.csv')
ratings['class'] = ratings['rating'].apply(classify)
ratings.sample(10)
```

## Задание №2.

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
           'Северо-Запад': ['петербург', 'псков', 'мурманск'],
           'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
len(geo_data['Центр'])
```

```python
def geo_data_classify(keyword):
    """
    
    Функция перебирает все комбинации вида регион - город и проверяет наличие названия города в запросе. 
    Если в поисковой строке содержится название города, то функция возвращает значение ключа, т.е. название региона.
    
    """
    for i in geo_data.keys():
        for j in range(len(geo_data[i])):
            if geo_data[i][j] in keyword:
                return i
    return 'undefined'
```

```python
kw = pd.read_csv(path_1 + 'keywords.csv')
kw['region'] = kw['keyword'].apply(geo_data_classify)
geo_filter = kw[ kw['region'] != 'undefined' ]
geo_filter.sample(10)
```

## Задание №3.

```python
# список из всех годов с 1950 по 2010

years = [ str(i) for i in range(1950, 2011, 1) ]
```

```python
movies = pd.read_csv(path_1 + 'movies.csv')
```

```python
def production_year(title):
    """
    Функция извлекающая год создания фильма из его названия.
    """
    for i in range(len(years)):
        if years[i] in title:
            return years[i]
        else:
            pass
    return '1900'
```

```python
# Добавляем столбик с годом создания фильма.
movies['year'] = movies['title'].apply(production_year)
# Оставляем фиильма из требуемого списка.
movies[ movies.year != '1900' ].sample(10)
```

```python
# Создаём словарь вида [movieId : rating]
rm = ratings.groupby('movieId').mean()['rating']
ratings_dict = rm.to_dict()
```

```python
# movies.join(rm).groupby('year').mean().sort_values('rating', ascending=0)
```

```python
def add_rating(row):
    """
    Вспомогательная функция для добавления столбика с рейтингом фильмов.
    Используется словарь rating_dict.
    """
    if row['movieId'] in ratings_dict.keys():
        return ratings_dict[row['movieId']]
```

```python
movies['rating'] = movies.apply(add_rating, axis=1)
movies.sample(10)
```

```python
# Групируем по годам и вычисляем среднюю величину рейтинга.
movies.groupby('year').mean().sort_values('rating', ascending=0)['rating']
```
